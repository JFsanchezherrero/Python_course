# Some basic introduction to python functions
Let's review a few basic features of advanced python programming.

A function is a set of code that performs any given task, enabling the programmer to modularize a program. 

You create a function by using the `def` keyword. 
To call the function, use the function name, followed by parentheses.

All variables created in function definitions are local variables: they are known only to the function where are declared. 


```python
# define the function
def say_hello():
    print("Hi there!")
    
## call it
say_hello()
```

You usually want to provide something to become a variable

```python
# define the function
def say_hello(name):
    print("Hi there " + name + "!")
    
    
## call it
say_hello("Jose")
```

Set a default value for this variable

```python
# define the function
def say_hello(name="Fulanito"):
    print("Hi there " + name + "!")
    
    
## call it
say_hello("Jose")
say_hello()
```

Values can be any type of data type

```python
def sum_func(i1, i2):
    total = i1 + i2
    print(total)
    
sum_func(100, 51)
```

```python
def sum_func(i1, i2):
    total = i1 + i2
    print(total)
    
sum_func(1.2, 51)
```

```python
def sum_func(i1, i2):
    total = i1 + i2
    print(total)
    
sum_func("a", 1)
```

### Recursive functions

```python
def counter(num2count):
    if num2count<=0:
        print("Stop")
    else:
        print(num2count)
        counter(num2count-1)
        
        
counter(5)
```

Be careful with recursive functions!




## Examples of functions in real life

You use a function when you are going to repeat the same code many times and you want it to be reproducible (and you are lazy and prefer not to copy code again).


You can call functions to print some statements, but you usually want the function to return you a variable.

```python
def square(y):
    return(y*y)

print("The square of number 2 is " + str(square(2)))
```

Functions are stored in memory as a "variable" so you can call them any time.

```python
for i in range(11):
    print("The square of number %d is %d" %(i, square(i)))
```

1) Create a function that returns wether a number is odd or even. Call for the range between 0-20

Remenber that even numbers are divisible by 2 without remainders. They end in 0, 2, 4, 6, or 8. 

Odd numbers are not evenly divisible by 2 and end in 1, 3, 5, 7, or 9.

```python
# some more functions
def even_odd(n):
    if n%2 == 0:
        return "even"
    else:
        return "odd"

print("Number 7 is " + even_odd(7))
```

```python
for i in range(21):
    print("Number %d is %s" %(i, even_odd(i)))
```
