````python
### Code to test the different time when working with very big numbers
import sys, platform
import time
 
l=list(range(100000001))
t=tuple(range(100000001))
 
start = time.time_ns()
for i in range(len(t)):
    a = t[i]
end = time.time_ns()
print("Total lookup time for Tuple: ", end - start)
 
start = time.time_ns()
for i in range(len(l)):
    a = l[i]
end = time.time_ns()
print("Total lookup time for LIST: ", end - start)
````

## Control Flow Statements

In the vast majority of cases, we will have to manipulate our data and to do so we will use the concepts of iteration and logical operations. While logical operations allow us to compare values between variables (greater, less, equality) and the iteration visits the elements of a list, tuple, dictionary or any data structure that can be sequenced one by one.

### Logical Operations

Logical operations return either `True` or `False`, which are known as _boolean_ variables

````python
print(5 > 1)
````

````python
print("a" in "abcdef")
````

````python
print("g" in "abc")
````

````python
print(4 == 5)
````

### Conditional statements

We can control the flow of statements using the estructure `if...else` or `if...elif...else`. 

These type of structures are used to decide whether a condition is satisfied or not. Usually, logical operations are produced that return either `True` or `False` and determine the sense of the workflow

````python
a = 5
b = 6
if a > b:
    print('a is greater than b')
else:
    print('a is smaller or equal to b')
````

````python
a = 5
b = 5
if a > b:
    print('a is greater than b')
elif a < b:
    print('a is smaller than b')
else:
    print('a is equal to b')
````

### Iterations/Loops

Repetitive statements or looping statements. 
It is used when the program requires a statements execution (or the repeated execution of a set of statements) until some condition for loop termination is satisfied.
There are two types of loops:
- For loop
- While loop

Further information [here](https://www.learnbyexample.org/python-for-loop/)


````python
# Use a list as an iterable
colors = ['red', 'green', 'blue', 'yellow']
for x in colors:
    print(x)
````

````python
S = 'python'
for x in S:
    print(x)
````

````python
# Generate a sequence of numbers from 0 6
for x in range(7):
    print(x)
````

````python
## Slicing a range from 2 to 6, every 2
for x in range(2, 7, 2):
    print(x)
````

````python
## modify a list while iterating
colors = ['red', 'green', 'blue']
for x in colors[:]:
    if x == 'red':
        colors.append('light-red')
    if x == 'blue':
        colors.append('light-blue')

print(colors)
````

Other type of iteration in python are generated using `while`. Check further details [here](https://realpython.com/python-while-loop/)

````python
## While loops
m = 5
i = 0
while i < m:
     print(i, end = ", ")
     i = i + 1 
##   i += 1
print("End")
````

````python
n = 5
while n > 0:
    n -= 1
    if n == 2:
        break
    print(n)
print('Loop ended.')
````

````python
n = 5
while n > 0:
    n -= 1
    if n == 2:
        continue
    print(n)
print('Loop ended.')
````

6.1) Iterate using `range()` from 0 to 200 and only print even numbers. Use either `if..else` or `continue`

````python
# use if ... else
for f in range(20):
    if f%2==0:
        print(f)
    else:
        pass
````

````python
## use continue
for f in range(20):
    if f%2>0:
        continue
    print(f)
````

6.2) Colors are visual perceptions that are identified by color categories such as: 
```
Black,Grey,Red,Blue,Orange,White,Brown,Pink,Yellow,Green,Purple...
```

We use the adjectives light, dark and bright to indicate the hue of the colors.

Iterate over the main categories of colors and add light, dark and bright to create a new list for each color categorie..

Expected output:

```
[ Black, light-Black, dark-Black, bright-Black ]
[ Grey, light-Grey, dark-Grey, bright-Grey ]
...
[ Purple, light-Purple, dark-Purple, bright-Purple ] 
```


````python
color_cat = "Black,Grey,Red,Blue,Orange,White,Brown,Pink,Yellow,Green,Purple"
color_hue = ['light', 'dark','bright']
````

````python
color_cat_list = color_cat.split(",")
color_cat_list
````

````python
## iterate using color_cat_list
for col in color_cat_list:
    
    ## create a new list
    my_list = []
    
    ## append color
    my_list.append(col)
    
    ## iterate over color hue items
    for hue in color_hue:
        ## add elements
        my_list.append(hue + '-' + col)
    
    ## print list for each color
    print(my_list)
    
    
````
