# Some basic introduction to python: dictionaries
Let's review a few basic features of advanced python programming data types.

## Python dictionaries

**_Dictionaries_** are used to store data values in key:value pairs.

A dictionary is a collection which is ordered*, changeable and do not allow duplicates. Dictionary elements are accessed via keys.

Additional details [here](https://realpython.com/python-dicts/) or [there](https://www.w3schools.com/python/python_dictionaries.asp)

### <i class="fa fa-search"></i> Exemple

````python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

print(thisdict)
````

````python
## Access a pair of values by its key 
print(thisdict["brand"])
````

````python
## Access a pair of values using get
print(thisdict.get("brand"))
````

Dictionaries can store any type of data type

### <i class="fa fa-search"></i> Exemple

````python
thisdict = {
  "brand": "Ford",
  "electric": False,
  "year": 1964,
  "height": 1.75,
  "colors": ["red", "white", "blue"],
  10 : "A number can also be the key"
}
print(thisdict)
````

Duplicate values will overwrite existing values:
### <i class="fa fa-search"></i> Exemple

````python
## check how duplicate values is overwritten
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964,
  "year": 2020
}
print(thisdict)
````

### Creating a dictionary
There are following three ways to create a dictionary.
- Using curly brackets: The dictionaries are created by enclosing the comma-separated Key: Value pairs inside the `{}` curly brackets. The colon `:` is used to separate the key and value in a pair.
- Using `dict()` constructor:  Create a dictionary by passing the comma-separated key: value pairs inside the `dict()`.
- Using sequence having each item as a pair (key-value)

### <i class="fa fa-search"></i> Exemple

````python
# An empty dictionary
dict_test = dict()
````

````python
# create a dictionary using {}
person = {"name": "Jessa", "country": "USA", "telephone": 1178}
print(person)
````

````python
# create a dictionary using dict()
person = dict({"name": "Jessa", "country": "USA", "telephone": 1178})
print(person)
````

````python
# create a dictionary from sequence having each item as a pair
person = dict([("name", "Jessa"), ("country", "USA"), ("telephone", 1178)])
print(person)
````

Use the following dictionary methods to retrieve all key and values at once

| Method	| Description|

| values()	| Returns the list of all values present in the dictionary|

### Iterating a dictionary
We can iterate through a dictionary using a for-loop and access the individual keys and their corresponding values. 
Let us see this with an example.

### <i class="fa fa-search"></i> Example 1

````python
# Iterating the dictionary using for-loop
print('key', ':', 'value')
for key in MLB_team:
    print(key, ':', MLB_team[key])
````

### <i class="fa fa-search"></i> Example 2

````python
# using items() method
print('key', ':', 'value')
for key_value in MLB_team.items():
    # first is key, and second is value
    print(key_value[0], key_value[1])
````

### Nested dictionaries
This are dictionaries that have one or more dictionaries as their members. It is a collection of many dictionaries in one dictionary.

### <i class="fa fa-search"></i> Example

````python
# address dictionary to store person address
address = {"state": "Texas", 'city': 'Houston'}

# dictionary to store person details with address as a nested dictionary
person = {'name': 'Jessa', 'company': 'Google', 'address': address}

# Display dictionary
print("person:", person)
````

````python
# Get nested dictionary key 'city'
print("City:", person['address']['city'])
````

````python
# Iterating outer dictionary
print("Person details")
for key, value in person.items():
    if key == 'address':
        # Iterating through nested dictionary
        print("Person Address")
        for nested_key, nested_value in value.items():
            print(nested_key, ':', nested_value)
    else:
        print(key, ':', value)
````
